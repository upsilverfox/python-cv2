```python
import cv2
import numpy as np
img = np.zeros((3,3),dtype=np.uint8)
print(img)
print(img.shape)
img = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR) 
print(img)
print(img.shape)
cv2.namedWindow("Image")
cv2.imshow("Image",img)
cv2.waitKey(0)
```
